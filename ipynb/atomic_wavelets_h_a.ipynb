```
import pandas as pd
import math
import cmath
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
%pylab inline
def ha(x,a):
  result=[]
  kmax=80
  mmax=15
  for i in x:
    if (i<=-1/(a-1)) or (i>=1/(a-1)):
      result.append(0)
    else:
      S=0.5
      for k in range(1,kmax):
        F=1
        for m in range(1,mmax):
          p=((a-1)*math.pi*k)/(a**m)
          F*=math.sin(p)/p
        S+=F*math.cos((a-1)*math.pi*k*i)
      result.append((a-1)*S)
  return(result)
def derha(x,a):
  der=(0.5*a**2)*(np.array(ha(a*x+1,a))-np.array(ha(a*x-1,a)))
  return(der)
def fi(w,a):
  fi=(np.array(ha(3*w/(2*math.pi)-1,a))+np.array(ha(3*w/(2*math.pi),a))+np.array(ha(3*w/(2*math.pi)+1,a)))**0.5
  return(fi)
def H0(w,a):
  H0=np.array(fi(2*(w-2*math.pi),a))+np.array(fi(2*w,a))+np.array(fi(2*(w+2*math.pi),a))
  return(H0)
def psi(w,a):
  psi=abs(np.exp(-1j*w/2)*(np.array(fi(w-2*math.pi,a))+np.array(fi(w+2*math.pi,a)))*fi(w/2,a))
  return(psi)
def deltader(x,a):
  dydx=np.diff(ha(x,a))/np.diff(x)
  dydx=np.append(0,dydx)
  pylab.grid(True)
  plt.plot(x,(dydx-derha(x,a)),'g--')
  pylab.xlabel('x')
  pylab.ylabel('f(x)')
def integral2(f, x, a, b):
  N = 100
  dw = (b-a)/N
  wi = np.arange(a, b+dw, dw)
  summa = []
  for k in range(len(x)):
    summa.append(sum(f(wi,x[k])*dw))
  return(np.array(summa))
def fio(w, x):
  return(fi(w,a)*np.cos(w*x))
def fii(x):
  return(integral2(fio, x, 0, 4*pi/3)/pi)
def psio(w, x):
  return(fi(w/2,a)*fi(w-2*math.pi,a)*np.cos(w*(x+0.5)))
def psix(x):
  return(integral2(psio, x, 2*math.pi/3, 8*math.pi/3)/math.pi)
```
